# ReAct Agent using Google Gemini API

This notebook demonstrates a simple implementation of a REACT (Reasoning + Acting) AI agent.

The agent:
- Takes user input (query)
- Uses Gemini API to generate structured reasoning
- Follows REACT format:
  - Thought
  - Action
  - Observation
  - Final Answer

This is useful for building:
- AI assistants
- Autonomous agents
- Decision pipelines

```python
import google.generativeai as genai
```

Explanation

- Imports Google’s Gemini AI

- This library allows interaction with Gemini models (like GPT but from Google)

```python
genai.configure(api_key="YOUR_API_KEY")
```

Explanation

- Connects your code to Gemini API

- API key is required for authentication

```python
model = genai.GenerativeModel("gemini-3.1-pro-preview")
```

Explanation

- Initializes the AI model

- "gemini-3.1-pro-preview" is a powerful reasoning model

- This model will generate responses for our agent

```python
def react_agent(query):
    prompt = f"""
You are a REACT (Reasoning + Acting) AI agent.

Follow this exact format:

Thought: think step by step
Action: what action to take
Observation: result of action
Final Answer: final response

Question: {query}
"""

    response = model.generate_content(prompt)
    return response.text
```

### Explanation

This function implements the REACT agent logic.

#### Step-by-step:

1. The user query is passed as input.
2. A structured prompt is created using the REACT format.
3. The prompt forces the model to:
   - Think step-by-step (Thought)
   - Decide an action (Action)
   - Describe results (Observation)
   - Provide final output (Final Answer)

4. The prompt is sent to the Gemini model using:
   model.generate_content()

5. The generated response is returned as text.

#### Key Concept:
This is an example of **prompt engineering**, where the behavior of the AI is controlled through structured instructions.

```python
while True:
    q = input("Ask (type 'exit' to quit): ")
    if q.lower() == "exit":
        break
    print(react_agent(q))
```

Explanation

- Runs infinite loop → keeps asking user for input until you type exit

- Calls react_agent() for every query
