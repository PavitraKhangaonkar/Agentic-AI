# Tree of Thought (ToT) using Gemini API

## Objective
To implement the Tree of Thought (ToT) reasoning framework using a Large Language Model (Gemini) and visualize reasoning paths using Graphviz.

## Description

This notebook demonstrates:
- Generating multiple reasoning paths using LLM
- Evaluating and selecting the best reasoning
- Visualizing reasoning as a tree structure

## Key Concept

Tree of Thought extends traditional reasoning:

- Chain-of-Thought → Single path
- Tree-of-Thought → Multiple paths + evaluation

This improves decision-making for complex problems.


````python
import google.generativeai as genai
from graphviz import Digraph
````

### Explanation

- `google.generativeai` → Used to access Gemini LLM
- `graphviz.Digraph` → Used to visualize reasoning as a tree

Graph visualization is important because:
- It shows branching decisions
- Helps understand reasoning flow

````python
# 🔑 Setup API
genai.configure(api_key="YOUR_API_KEY")

model = genai.GenerativeModel("gemini-3.1-pro-preview")
````

### Explanation

- API key connects your code to Gemini services
- `GenerativeModel` initializes the LLM

````python
# 🌳 Generate thoughts
def generate_thoughts(problem, num_thoughts=3):
    prompt = f"""
You are solving a complex problem using Tree of Thought reasoning.

Problem: {problem}

Generate {num_thoughts} different reasoning approaches.

Format:
Thought 1:
Thought 2:
Thought 3:
"""
    response = model.generate_content(prompt)
    return response.text
````

### Thought Generation

- This function generates multiple reasoning paths using Gemini.
- Each "thought" represents a possible approach.

Key Idea:
Instead of one solution → generate multiple strategies

LLM Role:
- Acts as a "thought generator"
- Produces diverse reasoning paths

````python
# 🧠 Evaluate thoughts
def evaluate_thoughts(problem, thoughts):
    prompt = f"""
Problem: {problem}

Here are multiple reasoning approaches:

{thoughts}

Select the best solution and explain why.
"""
    response = model.generate_content(prompt)
    return response.text
````

### Thought Evaluation

- This function selects the best reasoning path.
- Gemini evaluates:
  - effectiveness
  - feasibility
  - clarity

Key Concept:
This replaces random scoring with LLM-based reasoning.

This step is critical because:
→ It decides which branch to follow

````python
# 🌳 Visualize tree
def visualize_tree(problem, thoughts):
    dot = Digraph()

    dot.node("Problem", problem)

    thought_list = thoughts.split("Thought")

    for i, t in enumerate(thought_list[1:], start=1):
        node_name = f"T{i}"
        dot.node(node_name, t.strip())
        dot.edge("Problem", node_name)

    dot.render("tree_of_thought", format="png", view=True)
````

### Visualization

- Creates a tree structure using Graphviz
- Root node → Problem
- Child nodes → Thought branches

Steps:
1. Split thoughts into list
2. Create nodes for each thought
3. Connect them to root

Output:
- Generates a PNG image showing reasoning tree

Benefit:
→ Makes AI reasoning interpretable

````python
# 🚀 Run
if __name__ == "__main__":
    problem = "How to reduce traffic congestion in a city?"

    thoughts = generate_thoughts(problem)
    print("\nGenerated Thoughts:\n", thoughts)

    evaluation = evaluate_thoughts(problem, thoughts)
    print("\nBest Solution:\n", evaluation)

    visualize_tree(problem, thoughts)
````

### Execution Flow

1. Define problem
2. Generate multiple thoughts
3. Evaluate best solution
4. Visualize reasoning tree

Pipeline:

Problem → Thought Generation → Evaluation → Visualization

This is a basic implementation of Tree-of-Thought reasoning.

## System Architecture

### Components

1. Input Layer
   - User problem

2. LLM Layer (Gemini)
   - Thought generation
   - Thought evaluation

3. Reasoning Layer
   - Multi-path exploration

4. Visualization Layer
